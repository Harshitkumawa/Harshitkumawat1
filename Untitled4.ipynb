```python
'''
Created on April 25, 2024

@author: Rushikesh
'''

import turtle  # allows us to use the turtles library

wn = turtle.Screen()  # creates a graphics window
t=turtle.Turtle()
h=turtle.Turtle()
p=turtle.Turtle()
c=turtle.Turtle()
s=turtle.Turtle()
g=turtle.Turtle()
w=turtle.Turtle()

def create_square(t):
    """Write your function defination here and tell us what
     it do?"""
    t.penup()
    t.goto(-200, 200)  # Move the turtle to the starting position
    t.pendown()
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
    t.left(90)
    t.forward(100)
def create_triangle(h):
    """Write your function defination here and tell us what
     it do?"""
    h.penup()
    h.goto(100, 170)  # Move the turtle to the starting position
    h.pendown()
    h.forward(100)
    h.left(120)
    h.forward(100)
    h.left(120)
    h.forward(100)
def create_penta(p):
    """Write your function defination here and tell us what
     it do?"""
    p.penup()
    p.goto(150, 0)
    p.pendown()
    
    # Draw a pentagon
    p.forward(100)
    p.left(72)
    p.forward(100)
    p.left(72)
    p.forward(100)
    p.left(72)
    p.forward(100)
    p.left(72)
    p.forward(100)
def create_circle(c):
    """Write your function defination here and tell us what
     it do?"""
    c.penup()
    c.goto(10, 10)  # Move the turtle to the starting position
    c.pendown()
    c.circle(100)
def create_star(s):
    """Write your function defination here and tell us what
     it do?"""
    s.penup()
    s.goto(30, 270)  # Move the turtle to the starting position
    s.pendown()
    i=0
    for i in range(5):
        s.forward(100)
        s.right(144)
def create_pat(g):
    """Write your function defination here and tell us what
     it do?"""
    g.penup()
    g.goto(-200, 0)
    g.pendown()
    i=0
    for i in range(100):
        g.forward(i)
        g.right(90)
def create_pat2(w):
    """Write your function defination here and tell us what
     it do?"""
    w.penup()
    w.goto(200, 0)
    w.pendown()
    for i in range(100):
        w.circle(i)
if __name__ == "__main__":
    def sketch():
        create_square(t)
        create_triangle(h)
        create_penta(p)
        create_circle(c)
        create_star(s)
        create_pat(g)
        create_pat2(w)
        wn.exitonclick()

sketch()
```
